# Penguin Exercise

```python
# load penguin data set (built in to seaborn)
# dataset named df
df =sns.load_dataset('penguins')

#display first 5 rows of df
df.head()
```

```python
# orig dataset has NaN - nulls
# if there is na drop entire row 
# normally this is too drastic
df = df.dropna(how='any')

# check shape (count rows, cols)
# confirm rows were dropped
df.shape
```

### Pivot table

```python
# any value can go in values just counting values
# fill na with 0 

pivot_df = pd.pivot_table(df, index='species', values ='body_mass_g', columns='island', aggfunc = 'count', fill_value=0)
pivot_df
```

```python
#pivot_df.plot(x='species', y='island', kind='bar)
#ax = pivot_df.plot(xticks='df.species.unique()', ylabel='island')

pivot_df.plot(kind='bar', ylabel='island')
```

```python
# obersarvations: 
# adelie: highest count, species spread across three islands
# genshoo only on Bisco
# needed to make pivot chart and then use pivot to make table - two steps
```

### Countplot

```python
#countplot - value counts for two categorical variables in one step

ax = sns.countplot(x="species", hue="island", data=df, palette="twilight")
ax
```

### BOX PLOTS / SWARM PLOTS

```python
sns.set(rc={'figure.figsize': (12,9)})
ax = sns.boxplot(x = 'island', y = 'body_mass_g', data = df, color = 'lightcyan').set(title='Penguin Body Mass By Island and sex')
ax = sns.swarmplot(x = 'island', y = 'body_mass_g', hue='sex', data = df, palette="twilight")
```

```python
ax = sns.boxplot(x = 'species', y = 'body_mass_g', data = df, color = 'lightcyan').set(title='Penguin Body Mass By Species and sex')
ax = sns.swarmplot(x = 'species', y = 'body_mass_g', hue='sex', data = df, palette = 'twilight')
```

```python
ax = sns.boxplot(x = 'sex', y = 'body_mass_g', data = df, color = 'lightcyan').set(title='Pengin Body Mass By Gender and species')
ax = sns.swarmplot(x = 'sex', y = 'body_mass_g', hue='species', data = df, palette = 'twilight')
```
